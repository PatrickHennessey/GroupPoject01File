```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Plot Graph for January-February 2014
# Create lists for 2014 graph

x_axis = range(1, 10)
theft_incidents = [345, 422, 404, 387, 420, 335, 338, 398, 435] 
assault_number = [254, 312, 288, 313, 335, 296, 305, 365, 323]
criminal_misch= [189, 229, 228, 193, 295, 211, 254, 273, 333]
vehicle_theft_incidents = [27, 43, 57, 37, 52, 32, 22, 35, 41]
dui_number = [37, 73, 117, 91, 126, 94, 72, 100, 94]
prost_number = [0, 0, 2, 0, 0, 0, 0, 1, 0]
    
```

```python
#Plot graph for 2014 and apply styling

theft, = plt.plot(x_axis, theft_incidents, color="red", label="Theft")
assault, = plt.plot(x_axis, assault_number, color="blue", label="Assault")
criminal_mischief, = plt.plot(x_axis, criminal_misch, color="green", label="Criminal Mischief")
vehicle_theft, = plt.plot(x_axis, vehicle_theft_incidents, color="purple", label="Vehicle Theft")
driving_influence, = plt.plot(x_axis, dui_number, color="orange", label="Impaired Driving")
prostitution, = plt.plot(x_axis, prost_number, color="magenta", label="Prostitution")

#Add labels to graph
plt.title("New York Crime Snap Shot Jan-Feb 2014 (Superbowl)")
plt.xlabel("First Nine Weekends 2014")
plt.ylabel("Number of Incidents")

plt.legend(handles=[theft, assault, criminal_mischief, vehicle_theft, driving_influence, prostitution], loc="best")

plt.savefig("NewYork2014.png")
```
