```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#Read crime csv
ny_df = pd.read_csv("NY_Crime_Data_Filtered_New.csv")
ny_df.head()
```

```python
#Converting date to datetime
ny_df['CMPLNT_FR_DT'] = pd.to_datetime(ny_df['CMPLNT_FR_DT'])
```

```python
# Combining types of crimes
ny_df = ny_df.replace({"OFF.AGNST PUB ORD SENSBLTY &": "Criminal Mischief", "CRIMINAL MISCHIEF & RELATED OF": "Criminal Mischief",\
                      "DISORDERLY CONDUCT": "Criminal Mischief", "OFFENSES AGAINST PUBLIC SAFETY": "Criminal Mischief", "PETIT LARCENY":\
                      "Theft", "INTOXICATED & IMPAIRED DRIVING": "Impaired Driving", "GRAND LARCENY OF MOTOR VEHICLE": "Vehicle Theft",\
                      "ASSAULT 3 & RELATED OFFENSES": "Assault", "PROSTITUTION & RELATED OFFENSES": "Prostitution"})

ny_df.head(25)
```

```python
# Passing OFNS_DESC items into series to be filtered out
ny_df_filtered = ny_df[ny_df.OFNS_DESC.isin(["Criminal Mischief", "Theft", "Impaired Driving", "Prostitution", "Assault",\
                                            "Vehicle Theft"])]
ny_df_filtered.tail(25)
```

```python
#Filtering dates for first graph: Jan-Feb 2013
ny_df_2013_filter = ny_df_filtered[ny_df_filtered['CMPLNT_FR_DT']>="01/01/2013"]
ny_df_2013_filter.tail()
```

```python
# Filtering date: nothing after Feb 28, 2013
ny_df_feb_2013 = ny_df_2013_filter[ny_df_2013_filter['CMPLNT_FR_DT']<="02/28/2013"]
ny_df_feb_2013.head()
```
