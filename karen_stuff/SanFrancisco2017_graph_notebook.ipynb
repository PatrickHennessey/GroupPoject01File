```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Create Lists for San Francisco 2017

x_axis2017 = range(1, 10)
theft_incidents2017 = [267, 254, 266, 292, 307, 242, 261, 314, 246] 
assault_number2017 = [100, 72, 82, 72, 71, 94, 84, 68, 91]
drunk_disorder2017 = [70, 64, 63, 83, 51, 67, 68, 73, 57]
vehicle_theft_incidents2017 = [30, 22, 30, 36, 47, 40, 44, 45, 41]
drug_narc2017 = [12, 16, 30, 20, 21, 16, 10, 10, 23]
dui_number2017 = [5, 3, 2, 4, 8, 3, 1, 6, 4]
prost_number2017 = [0, 0, 0, 0, 0, 0, 2, 0, 0]
```

```python
# Graph for San Francisco 2017

theft, = plt.plot(x_axis2017, theft_incidents2017, color="red", label="Theft")
assault, = plt.plot(x_axis2017, assault_number2017, color="blue", label="Assault")
drunk_disorderly, = plt.plot(x_axis2017, drunk_disorder2017, color="green", label="Drunk/Disorderly")
vehicle_theft, = plt.plot(x_axis2017, vehicle_theft_incidents2017, color="purple", label="Vehicle Theft")
drug_narcotics, = plt.plot(x_axis2017, drug_narc2017, color="yellow", label="Drugs/Narcotics")
driving_influence, = plt.plot(x_axis2017, dui_number2017, color="orange", label="DUI")
prostitution, = plt.plot(x_axis2017, prost_number2017, color="magenta", label="Prostitution")

#Add labels to graph
plt.title("San Francisco Crime Snap Shot Jan-Feb 2017")
plt.xlabel("First Nine Weekends 2017")
plt.ylabel("Number of Incidents")

plt.legend(handles=[theft, assault, drunk_disorderly, vehicle_theft, drug_narcotics, driving_influence, prostitution], loc="best")

plt.savefig("SanFrancisco2017.png")
```
