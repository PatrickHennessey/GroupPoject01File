```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
# Plot Graph for January-February 2013
# Create lists for 2013 graph

x_axis = range(1, 10)
theft_incidents = [373, 422, 408, 341, 362, 313, 354, 391, 323] 
assault_number = [262, 308, 290, 310, 297, 259, 304, 317, 279]
criminal_misch= [272, 276, 254, 168, 231, 166, 219, 228, 246]
vehicle_theft_incidents = [29, 42, 37, 33, 38, 14, 26, 33, 30]
dui_number = [62, 84, 75, 80, 77, 39, 91, 93, 86]
prost_number = [0, 0, 0, 0, 1, 0, 0, 0, 0]
    
```

```python
#Plot graph for 2013 and apply styling

theft, = plt.plot(x_axis, theft_incidents, color="red", label="Theft")
assault, = plt.plot(x_axis, assault_number, color="blue", label="Assault")
criminal_mischief, = plt.plot(x_axis, criminal_misch, color="green", label="Criminal Mischief")
vehicle_theft, = plt.plot(x_axis, vehicle_theft_incidents, color="purple", label="Vehicle Theft")
driving_influence, = plt.plot(x_axis, dui_number, color="orange", label="Impaired Driving")
prostitution, = plt.plot(x_axis, prost_number, color="magenta", label="Prostitution")

#Add labels to graph
plt.title("New York Crime Snap Shot Jan-Feb 2013")
plt.xlabel("First Nine Weekends 2013")
plt.ylabel("Number of Incidents")

plt.legend(handles=[theft, assault, criminal_mischief, vehicle_theft, driving_influence, prostitution], loc="best")

plt.savefig("NewYork2013.png")
```
