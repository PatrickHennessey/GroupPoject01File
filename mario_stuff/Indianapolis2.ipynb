```python
%matplotlib notebook

# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_axis = range(1,10)
stolen_vehicle_number = [16,21,27,19,32,26,22,12,24]
assault_other_weapon_number = [15,17,1,9,2,12,10,9,15]
assault_knife_number = [4,6,2,2,6,3,4,5,2]
```

```python
stolen_vehicle, = plt.plot(x_axis, stolen_vehicle_number, color="red", label="Stolen Vehicle")
other_weapon, = plt.plot(x_axis,assault_other_weapon_number, color="blue", label="Assault o. Weapon")
assault_knife, = plt.plot(x_axis,assault_knife_number, color="green", label="Assault Knife")

plt.title("Indianapolis Snap Shot Jan-Feb 2013")
plt.xlabel("First 9 Weekends 2013")
plt.ylabel("Number of Incidents")

plt.legend(handles=[stolen_vehicle, other_weapon, assault_knife], loc="best")
plt.savefig("Indianapolis2013.png")
```
